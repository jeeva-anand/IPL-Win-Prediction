```python
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import sys

sys.path.append('../../IPL Team Win Prediction')

finally_df = pd.read_csv('../data/preprocessed/final_df.csv')

finally_df.head()
```

```python

data = finally_df.copy()

test = data.result

data.drop('result', axis=1)

xtrain, xtest,ytrain,ytest = train_test_split(data,test,test_size=0.3,random_state=42)
```

```python
data.columns
```

```python

cf = ColumnTransformer(transformers=
    [('tnf1', OneHotEncoder(sparse_output=False,drop='first'),['batting_team','city','bowling_team'])] ,remainder ='passthrough')
```

```python

pipe1 = Pipeline(steps=[
    ('step1' , cf),
    ('step2' , LogisticRegression(solver='liblinear'))
])
```

```python
pipe1.fit(xtrain,ytrain)
ypred = pipe1.predict(xtest)

score = accuracy_score(ytest,ypred)

score
```
